```python
data1['label'] = data1.n_packets == data1.n_normal
data2['label'] = data2.n_packets == data2.n_normal
data3['label'] = data3.n_packets == data3.n_normal
data4['label'] = data4.n_packets == data4.n_normal
data5['label'] = data5.n_packets == data5.n_normal
data6['label'] = data6.n_packets == data6.n_normal
```

```python
data1.to_parquet('../../data/pcap_files_csv/capture20110810.parquet_1.0s.parquet')
data2.to_parquet('../../data/pcap_files_csv/capture20110811.parquet_1.0s.parquet')
data3.to_parquet('../../data/pcap_files_csv/capture20110815.parquet_1.0s.parquet')
data4.to_parquet('../../data/pcap_files_csv/capture20110817.parquet_1.0s.parquet')
data5.to_parquet('../../data/pcap_files_csv/capture20110818.parquet_1.0s.parquet')
data6.to_parquet('../../data/pcap_files_csv/capture20110818-2.parquet_1.0s.parquet')
```
